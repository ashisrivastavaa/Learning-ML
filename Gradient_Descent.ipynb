```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Generata data using dummy function
x=np.arange(10)
y=(x-5)**2+3
print(x,y)
```

```python
 #Visualize
plt.style.use("seaborn-v0_8") #plot the grid (graph like background)
plt.plot(x,y)  #plot the data point on x-y plane
plt.xlabel("X") #label x axis with X
plt.ylabel("Y")  #label y axis with Y
plt.show()  #display in output
```

```python
#Gradient Descent
X=0  #initialize X with any random value
learning_rate=0.1 #learning rate(if increase the learning rate the jump size is higher/bigger and also it diverge if learning rate is very large)
plt.plot(x,y)
for i in range(50):
    grad= 2*(X-5) #gradient (dy/dx)
    X=X-learning_rate*grad #update X
    Y=(X-5)**2+3 #calculate Y
    plt.scatter(X,Y) #plot the data point(see the dot)
    print(X)
plt.show()
```

```python
#Gradient Descent
X=10  #initialize  x to last
learning_rate=0.1 #learning rate
plt.plot(x,y)
for i in range(50):
    grad= 2*(X-5) #gradient (dy/dx)
    X=X-learning_rate*grad #update X
    Y=(X-5)**2+3 #calculate Y
    plt.scatter(X,Y) #plot the data point(see the dot)
    print(X)
plt.show()
```
